```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
data = {'timesPerDay': [0, 1, 2, 3, 4, 5, 6],
        'count': [85, 18, 27, 20, 20, 2, 3]}
df = pd.DataFrame(data)

# Plot Bubble Chart
plt.scatter(df['timesPerDay'], df['count'], s=df['count']*10, c=df['count'], alpha=0.5)
plt.xlabel('timesPerDay')
plt.ylabel('count')
plt.title('Bubble Chart for number of times they drink per day ')
plt.colorbar()
plt.show()
```

```python
y = np.array([85, 18, 27,20, 20, 2,3 ])
mylabels = ["ZERO","ONE","TWO" ," THREE","FOUR", "FIVE", "SIX"]

myexplode = [0.2, 0, 0, 0,0,0,0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.title("how many times do you drink energy drinks")
plt.show()
```
