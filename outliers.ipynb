## Z SCORE TO FIND OUTLIERS

```python
for col in df.columns:
  col_zscore = col + '_zscore'

  df[col_zscore]= (df[col]-df[col].mean())/df[col].std(ddof=0)
```

```python
df['outlier'] = (abs(df['Test_Score_zscore'])>3).astype(int)
```

```python
df.drop(columns=['Student_ID_zscore'],inplace=True)
```

## USING INTERQUARTILE RANGE TO DETECT OUTLIERS

```python
#df
```

```python
Q1= df.Test_Score.quantile(0.25)
Q3= df.Test_Score.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3-Q1
```

```python
lower_limit = Q1-1.5*IQR
upper_limit = Q3+1.5*IQR
lower_limit, upper_limit
```

```python
df[(df.Test_Score<lower_limit)|(df.Test_Score> upper_limit) ]
```

```python
df_no_outliers = df[(df.Test_Score<lower_limit)|(df.Test_Score> upper_limit) ]
df_no_outliers
```
