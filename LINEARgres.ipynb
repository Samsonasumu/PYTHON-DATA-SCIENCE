```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/lot3_cleaned_outliers_removed.csv")
df.head()
```

** **DATA CLEANING**

```python
df['shipping']
```

```python
import re

df['shipping'] = df['shipping'].apply(lambda x: re.sub('[^\d.]', '', x))
df['trending'] = df['trending'].apply(lambda x: re.sub('[^\d.]', '', x))


df.head()
```

```python
df['shipping'] = df['shipping'].fillna(0)
```

```python
df['shipping'] = df['shipping'].replace("",0)

df['trending'] = df['trending'].replace("",0)

```

```python
df
```

```python
#remove the  K sign() in the seller_item_sold column
df['seller_item_sold'] = df['seller_item_sold'].str.replace("K", ",000")
df['seller_item_sold'] = df['seller_item_sold'].str.replace("M", "000,000")

```

```python
df
```
