```python
import pandas as pd
import requests

import warnings
warnings.filterwarnings('ignore')
```

```python
response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=4666f0ddc04ac1eaf22895........&language=en-US&page=1")
```

```python
response.json()['results']
```

```python
tempo_df=pd.DataFrame(response.json()['results'])[['id','popularity','release_date','title','vote_average','vote_count']]
```

```python
#creating an empty dataframe to store our final dataset
df_final = pd.DataFrame()
```

```python
df_final
```

```python
#we needto run a for loop since we have 555 pages
#we need to iterATE UNITL 556
for k in range(1,556):
    response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=4666f0ddc04ac1eaf22895a5a6f63779&language=en-US&page={}".format(k))
    tempo_df =pd.DataFrame(response.json()['results'])[['id','popularity','release_date','title','vote_average','vote_count']]
    df_final=df_final.append(tempo_df,ignore_index=True)
```
